```python
from pyspark.sql import SparkSession
from pyspark.sql import Row
from pyspark.sql.functions import *
from pyspark.sql.types import *

working_directory = 'main/jars/'

spark = SparkSession \
    .builder \
    .appName('data-public_local_2') \
    .config('spark.driver.extraClassPath',
            working_directory + 'postgresql-42.2.5.jar') \
    .master("local[*]") \
    .getOrCreate()
```

```python
# retrieves data from parket files
output_directory = 'main/outputdata'

badges = spark.read.parquet(output_directory + '/Badges/*')
comments = spark.read.parquet(output_directory + '/Comments/*')
posts = spark.read.parquet(output_directory + '/Posts/*')
post_history = spark.read.parquet(output_directory + '/PostHistory/*')
post_links = spark.read.parquet(output_directory + '/PostLinks/*')
users = spark.read.parquet(output_directory + '/Users/*')
votes = spark.read.parquet(output_directory + '/Votes/*')
tags = spark.read.parquet(output_directory + '/Tags/*')
```

```python
# some query
comments.select('*').where(col("CreationDate") < "2011-05-24T19:39:08.653").show()
```

```python
# print data
print("\n\n++++++++++++++++  BADGES  ++++++++++++++++\n")
badges.select("*").show()
print("\n\n++++++++++++++++  COMMENTS ++++++++++++++++\n")
comments.select("*").show()
print("\n\n++++++++++++++++  POSTS ++++++++++++++++\n")
posts.select("*").show()
print("\n\n++++++++++++++++  POST_HISTORY ++++++++++++++++\n")
post_history.select("*").show()
print("\n\n++++++++++++++++  USERS ++++++++++++++++\n")
users.select("*").show()
print("\n\n++++++++++++++++  VOTES ++++++++++++++++\n")
votes.select("*").show()
print("\n\n++++++++++++++++  TAGS  ++++++++++++++++\n")
tags.select("*").show()
```
